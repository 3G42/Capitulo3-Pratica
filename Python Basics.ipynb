```python
### Contador

contador = 0;

while contador <= 9:
  contador = contador + 1
  if contador == 1:
    print(contador, 'item limpo')
  else:
    print(contador,"itens limpos")
print('Fim da repetição while')
```

```python
contador = 0

while True:
  if contador < 10:
    contador = contador + 1
    if contador == 1:
      print(contador, 'item limpo')
    elif contador == 9:
      print(contador,"itens limpos")
      print('Falta 1 item')
    else:
      print(contador,"itens limpos")
  else:
    break
```
